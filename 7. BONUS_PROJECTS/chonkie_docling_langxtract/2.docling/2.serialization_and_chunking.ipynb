# 5. Custom Picture Serializer with Annotations

## Concept Overview
Custom picture serializers can extract and embed image annotations (descriptions, classifications, etc.) directly into the serialized text. This enriches the output with metadata about images.

````python
from typing import Any, Optional
from docling_core.transforms.serializer.base import (
    BaseDocSerializer,
    SerializationResult,
)
from docling_core.transforms.serializer.common import create_ser_result
from docling_core.transforms.serializer.markdown import (
    MarkdownPictureSerializer,
)
from docling_core.types.doc.document import (
    DoclingDocument,
    ImageRefMode,
    PictureItem,
)
from typing_extensions import override

class AnnotationPictureSerializer(MarkdownPictureSerializer):
    """Custom picture serializer that includes image annotations"""
    
    @override
    def serialize(
        self,
        *,
        item: PictureItem,
        doc_serializer: BaseDocSerializer,
        doc: DoclingDocument,
        separator: Optional[str] = None,
        **kwargs: Any,
    ) -> SerializationResult:
        text_parts: list[str] = []

        # Reuse existing result
        parent_res = super().serialize(
            item=item,
            doc_serializer=doc_serializer,
            doc=doc,
            **kwargs,
        )
        text_parts.append(parent_res.text)

        # Append annotations if available
        if item.meta is not None and item.meta.description is not None:
            text_parts.append(
                f"<!-- Picture description: {item.meta.description.text} -->"
            )

        text_res = (separator or "\n").join(text_parts)
        return create_ser_result(text=text_res, span_source=item)

# Use custom serializer
serializer = MarkdownDocSerializer(
    doc=doc,
    picture_serializer=AnnotationPictureSerializer(),
    params=MarkdownParams(
        image_mode=ImageRefMode.PLACEHOLDER,
        image_placeholder="",
    ),
)
ser_result = serializer.serialize()
annotated_text = ser_result.text

console.print(Panel(annotated_text[:400] + "...", title="Custom Picture Serializer with Annotations"))
print(f"\nâœ“ Custom picture serializer applied")
print(f"âœ“ Annotations embedded in output")
````

# 6. Basic Hybrid Chunking

## Concept Overview
Hybrid chunking applies tokenization-aware refinements on top of document-based hierarchical chunking. It splits documents into chunks that respect token limits while preserving semantic structure.

````python
from docling.chunking import HybridChunker

# Convert markdown document for chunking
MD_SOURCE = "./mock_data_serialization/wiki_sample.md"
doc_md = DocumentConverter().convert(source=MD_SOURCE).document

# Basic hybrid chunking
chunker = HybridChunker()
chunk_iter = chunker.chunk(dl_doc=doc_md)

# Display first 3 chunks
print("ğŸ“‹ First 3 chunks:\n")
for i, chunk in enumerate(list(chunk_iter)[:3]):
    print(f"=== Chunk {i} ===")
    print(f"Text: {chunk.text[:150]}...")
    
    # Get contextualized text
    enriched_text = chunker.contextualize(chunk=chunk)
    print(f"Contextualized: {enriched_text[:150]}...")
    print()

print("âœ“ Basic hybrid chunking complete")
````

# 7. Hybrid Chunking with Contextualization

## Concept Overview
The `contextualize()` method enriches chunks with hierarchical context (e.g., section headers). This helps embeddings understand the chunk's position in the document structure.

````python
# Demonstrate contextualization
chunker = HybridChunker()
chunk_iter = chunker.chunk(dl_doc=doc_md)

print("ğŸ” Chunk vs Contextualized Comparison:\n")
for i, chunk in enumerate(list(chunk_iter)[:2]):
    print(f"=== Chunk {i} ===")
    print(f"Original:\n{chunk.text[:200]}\n")
    
    enriched = chunker.contextualize(chunk=chunk)
    print(f"Contextualized:\n{enriched[:200]}\n")
    print("-" * 80)

print("\nâœ“ Context enrichment adds document structure metadata")
````

# 8. Configuring HuggingFace Tokenizer

## Concept Overview
HuggingFace tokenizers ensure chunking aligns with embedding models. This prevents token overflow and maintains consistency between document processing and retrieval systems.

````python
from docling_core.transforms.chunker.tokenizer.huggingface import HuggingFaceTokenizer
from transformers import AutoTokenizer

# Configure with HuggingFace tokenizer
EMBED_MODEL_ID = "sentence-transformers/all-MiniLM-L6-v2"
MAX_TOKENS = 128  # Small for demonstration

tokenizer = HuggingFaceTokenizer(
    tokenizer=AutoTokenizer.from_pretrained(EMBED_MODEL_ID),
    max_tokens=MAX_TOKENS,
)

chunker = HybridChunker(
    tokenizer=tokenizer,
    merge_peers=True,
)

chunk_iter = chunker.chunk(dl_doc=doc_md)
chunks = list(chunk_iter)

# Display token counts
print(f"âš™ï¸ Tokenizer Config: {EMBED_MODEL_ID}")
print(f"Max tokens: {MAX_TOKENS}\n")

for i, chunk in enumerate(chunks[:5]):
    txt_tokens = tokenizer.count_tokens(chunk.text)
    ser_txt = chunker.contextualize(chunk=chunk)
    ser_tokens = tokenizer.count_tokens(ser_txt)
    print(f"Chunk {i}: {txt_tokens} tokens (raw) | {ser_tokens} tokens (contextualized)")

print(f"\nâœ“ Created {len(chunks)} chunks respecting token limits")
````

# 9. Token-Aware Chunk Merging

## Concept Overview
The `merge_peers` option merges undersized adjacent chunks when possible, reducing chunk count while staying within token limits. This improves efficiency for embedding and retrieval.

````python
# Compare with and without merging
MAX_TOKENS = 64

# Without merging
tokenizer_no_merge = HuggingFaceTokenizer(
    tokenizer=AutoTokenizer.from_pretrained(EMBED_MODEL_ID),
    max_tokens=MAX_TOKENS,
)
chunker_no_merge = HybridChunker(tokenizer=tokenizer_no_merge, merge_peers=False)
chunks_no_merge = list(chunker_no_merge.chunk(dl_doc=doc_md))

# With merging
chunker_merge = HybridChunker(tokenizer=tokenizer_no_merge, merge_peers=True)
chunks_merge = list(chunker_merge.chunk(dl_doc=doc_md))

print(f"ğŸ“Š Chunk Merging Comparison (max_tokens={MAX_TOKENS}):\n")
print(f"Without merging: {len(chunks_no_merge)} chunks")
print(f"With merging: {len(chunks_merge)} chunks")
print(f"Reduction: {len(chunks_no_merge) - len(chunks_merge)} chunks ({((len(chunks_no_merge) - len(chunks_merge)) / len(chunks_no_merge) * 100):.1f}%)")

print("\nâœ“ Merging reduces chunk count while respecting token limits")
````

# 11. Advanced Table Serialization - Default Strategy

## Concept Overview
Advanced chunking combines custom serialization providers with hybrid chunking. The default table strategy uses triplet format for better vector representation.

````python
from docling_core.transforms.chunker.hierarchical_chunker import (
    ChunkingDocSerializer,
    ChunkingSerializerProvider,
)
from docling_core.types.doc.labels import DocItemLabel

# Load or use existing doc
doc_for_advanced = doc  # Using the PDF doc from earlier

# Helper to find chunks with specific labels
def find_chunk_with_label(chunks, label: DocItemLabel):
    """Find first chunk containing specified label"""
    for i, chunk in enumerate(chunks):
        # Check if chunk has the label
        if hasattr(chunk, 'meta') and hasattr(chunk.meta, 'doc_items'):
            for item in chunk.meta.doc_items:
                if hasattr(item, 'label') and item.label == label:
                    return i, chunk
    return None, None

# Basic chunking with default serialization
tokenizer_advanced = HuggingFaceTokenizer(
    tokenizer=AutoTokenizer.from_pretrained(EMBED_MODEL_ID),
)
chunker_advanced = HybridChunker(tokenizer=tokenizer_advanced)
chunks_advanced = list(chunker_advanced.chunk(dl_doc=doc_for_advanced))

print(f"ğŸ” Advanced Table Serialization - Default Strategy\n")
print(f"Created {len(chunks_advanced)} chunks from document")

# Try to find a table chunk
pos, table_chunk = find_chunk_with_label(chunks_advanced, DocItemLabel.TABLE)
if table_chunk:
    ctx_text = chunker_advanced.contextualize(chunk=table_chunk)
    console.print(Panel(ctx_text[:500] + "...", title=f"Table Chunk (Position {pos})"))
else:
    print("No table chunks found in this document")

print("\nâœ“ Default table serialization uses triplet format")
````

# 14. Advanced Picture Serialization with Annotations

## Concept Overview
Advanced picture serializers extract metadata (descriptions, classifications, SMILES for molecules) and embed them in the chunked output. This is crucial for multimodal RAG systems.

````python
from docling_core.types.doc.document import (
    PictureClassificationData,
    PictureDescriptionData,
    PictureMoleculeData,
)

class AnnotationPictureSerializerAdvanced(MarkdownPictureSerializer):
    """Advanced picture serializer with full annotation support"""
    
    @override
    def serialize(
        self,
        *,
        item: PictureItem,
        doc_serializer: BaseDocSerializer,
        doc: DoclingDocument,
        **kwargs: Any,
    ) -> SerializationResult:
        text_parts: list[str] = []

        if item.meta is not None:
            # Add classification if available
            if item.meta.classification is not None:
                main_pred = item.meta.classification.get_main_prediction()
                if main_pred is not None:
                    text_parts.append(f"Picture type: {main_pred.class_name}")

            # Add molecule data if available
            if item.meta.molecule is not None:
                text_parts.append(f"SMILES: {item.meta.molecule.smi}")

            # Add description if available
            if item.meta.description is not None:
                text_parts.append(f"Picture description: {item.meta.description.text}")

        text_res = "\n".join(text_parts) if text_parts else "<!-- image -->"
        text_res = doc_serializer.post_process(text=text_res)
        return create_ser_result(text=text_res, span_source=item)

class ImgAnnotationSerializerProvider(ChunkingSerializerProvider):
    """Provider using annotation-based picture serializer"""
    def get_serializer(self, doc: DoclingDocument):
        return ChunkingDocSerializer(
            doc=doc,
            picture_serializer=AnnotationPictureSerializerAdvanced(),
        )

# Apply advanced annotation serializer
chunker_annotation = HybridChunker(
    tokenizer=tokenizer_advanced,
    serializer_provider=ImgAnnotationSerializerProvider(),
)

chunks_annotation = list(chunker_annotation.chunk(dl_doc=doc_for_advanced))
print(f"ğŸ¨ Advanced Picture Serialization with Annotations\n")

# Find picture chunk with annotations
pos, pic_chunk_ann = find_chunk_with_label(chunks_annotation, DocItemLabel.PICTURE)
if pic_chunk_ann:
    ctx_text = chunker_annotation.contextualize(chunk=pic_chunk_ann)
    console.print(Panel(ctx_text[:400] + "...", title=f"Annotated Picture Chunk (Position {pos})"))
else:
    print("No picture chunks found")

print("\nâœ“ Annotations provide rich metadata for multimodal search")
print("âœ“ Supports: classifications, descriptions, and molecule SMILES")
````

## Concept Overview
End-to-end pipeline from document conversion â†’ custom serialization â†’ hybrid chunking with token limits. This demonstrates a production-ready RAG ingestion pipeline.

````python
# Complete workflow demonstration
print("ğŸ”„ Complete Serialization & Chunking Workflow\n")
print("=" * 80)

# Step 1: Convert document
print("\n1ï¸âƒ£ Document Conversion")
source_doc = "./mock_data_serialization/sample_document.pdf"
converter = DocumentConverter()
document = converter.convert(source=source_doc).document
print(f"   âœ“ Converted: {source_doc}")

# Step 2: Configure custom serialization
print("\n2ï¸âƒ£ Configure Serialization")
class CustomSerializerProvider(ChunkingSerializerProvider):
    def get_serializer(self, doc):
        return ChunkingDocSerializer(
            doc=doc,
            table_serializer=MarkdownTableSerializer(),
            params=MarkdownParams(
                image_placeholder="[Image]",
            ),
        )
print("   âœ“ Custom table & image serializers configured")

# Step 3: Configure tokenization
print("\n3ï¸âƒ£ Configure Tokenization")
hf_tokenizer = HuggingFaceTokenizer(
    tokenizer=AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2"),
    max_tokens=256,
)
print(f"   âœ“ HuggingFace tokenizer: max_tokens=256")

# Step 4: Chunk with hybrid chunker
print("\n4ï¸âƒ£ Hybrid Chunking")
final_chunker = HybridChunker(
    tokenizer=hf_tokenizer,
    serializer_provider=CustomSerializerProvider(),
    merge_peers=True,
)
final_chunks = list(final_chunker.chunk(dl_doc=document))
print(f"   âœ“ Created {len(final_chunks)} chunks")

# Step 5: Display results
print("\n5ï¸âƒ£ Results Summary")
total_tokens = sum(hf_tokenizer.count_tokens(final_chunker.contextualize(chunk=c)) for c in final_chunks)
avg_tokens = total_tokens / len(final_chunks) if final_chunks else 0
print(f"   â€¢ Total chunks: {len(final_chunks)}")
print(f"   â€¢ Total tokens: {total_tokens}")
print(f"   â€¢ Average tokens/chunk: {avg_tokens:.1f}")

# Show first chunk
if final_chunks:
    print("\nğŸ“„ First Chunk Preview:")
    first_ctx = final_chunker.contextualize(chunk=final_chunks[0])
    console.print(Panel(first_ctx[:400] + "...", title="Chunk 0 (Contextualized)"))

print("\n" + "=" * 80)
print("âœ… Complete workflow executed successfully!")
````
